Q7 - Create a simple webpage with a card similar to question number 2. The card must be semi-transparent as
default and on hovering the card must be displayed clearly. Use CSS opacity to achieve this output ?

```python
-<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8">
  <meta name="viewport" content="width=device-width, initial-scale=1.0">
  <title>Document</title>
  <style>
    body {
      margin: 0;
      padding: 0;
      background-color: #f0f0f0; /* Light grey background */
    }

    .outer-div {
      background-color: #2E8B57; /* Classy matte green (SeaGreen) */
      width: 350px;
      margin: 30px auto 0; /* 30px from top, auto left-right, 0 bottom */
      padding: 20px;
      border-radius: 20px;
      text-align: center; /* Center image inside */
    }

    .outer-div img {
      width: 100%;
      height: auto;
      border-radius: 15px;
      display: block;
      transition: opacity 0.3s ease; /* Smooth transition for hover effect */
    }

    .outer-div:hover {
      opacity: 0.6;
      background-color: #4c8665 /* Slightly more transparent on hover */
    }
  </style>

</head>
<body>
  <div class="outer-div">
    <img src="" alt="Course Poster ">
    <p> <b>Course Name :</b> PW Full Stack Web Development </p>
  </div>
</body>
</html>
```
